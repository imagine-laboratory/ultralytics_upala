```python
path_directory = "H:/Datasets/VideoPineapple"
```

```python
video_dir1 = "DJI_20240308110454_0007_V_1"
video_dir2 = "DJI_20240308110013_0004_V_3"

video_combined = "DJI_0007_V_1_and_0004_V_3" # DJI_0008_V_1_and_0010_V_1

# Read to combined 
# DJI_20240308110958_0008_V_1
# DJI_20240308111117_0010_V_1
# DJI_20240308110958_0008_V_2
# DJI_20240308111117_0010_V_2
```

```python
import os

path_video_dir1_images = os.path.join(path_directory, video_dir1, "valid", "images")
path_video_dir1_labels = os.path.join(path_directory, video_dir1, "valid", "labels")

path_video_dir2_images = os.path.join(path_directory, video_dir2, "valid", "images")
path_video_dir2_labels = os.path.join(path_directory, video_dir2, "valid", "labels")

path_video_combined_images = os.path.join(path_directory, video_combined, "valid", "images")
path_video_combined_labels = os.path.join(path_directory, video_combined, "valid", "labels")
```

```python
def read_files(folder_path):
    # List all files and folders in the directory
    all_files_and_folders = os.listdir(folder_path)
    # Filter to include only files
    files = [f for f in all_files_and_folders if os.path.isfile(os.path.join(folder_path, f))]
    print(len(files))
    return files 
```

```python
import shutil
def copyfile(files_images_dir, path_video_combined_images, path_video_dir_images):
    for path_img in files_images_dir:

        # Copy the file with the new name
        path_combined_image = os.path.join(path_video_combined_images, path_img)
        path_image = os.path.join(path_video_dir_images, path_img)

        shutil.copy(path_image, path_combined_image)
    print("Finish")
```

```python
def copyfile_rename(files_images_dir, path_video_combined_images, path_video_dir_images, dest_count):

    count = dest_count
    mapping = {}
    for path_img in files_images_dir:
        ext = path_img.split(".")[1]
        new_img_name = f"frame_{count:05d}.{ext}"  # Increment and format as zero-padded (width=5)
        
        # Copy the file with the new name
        path_combined_image = os.path.join(path_video_combined_images, new_img_name)
        path_image = os.path.join(path_video_dir_images, path_img)

        shutil.copy(path_image, path_combined_image)
        count += 1
        mapping[path_img.split(".")[0]] = new_img_name.split(".")[0]
    return mapping

def copyfile_rename_labels(files_labels_dir, path_video_combined_labels, path_video_dir_labels, mapping):

    for path_img in files_labels_dir:
        ext = path_img.split(".")[1]
        namefile = path_img.split(".")[0]
        new_img_name = f"{mapping[namefile]}.{ext}"  # Increment and format as zero-padded (width=5)
        
        # Copy the file with the new name
        path_combined_image = os.path.join(path_video_combined_labels, new_img_name)
        path_image = os.path.join(path_video_dir_labels, path_img)

        shutil.copy(path_image, path_combined_image)
    print("Finish")
```

```python
os.makedirs(path_video_combined_images, exist_ok=True)
os.makedirs(path_video_combined_labels, exist_ok=True)

files_labels_dir1 = read_files(path_video_dir1_labels)
files_images_dir1 = read_files(path_video_dir1_images)
```

```python
copyfile(files_labels_dir1, path_video_combined_labels, path_video_dir1_labels)
copyfile(files_images_dir1, path_video_combined_images, path_video_dir1_images)
```

```python
files_labels_dir2 = read_files(path_video_dir2_labels)
files_images_dir2 = read_files(path_video_dir2_images)

mapping = copyfile_rename(files_images_dir2, path_video_combined_images, path_video_dir2_images, dest_count=len(files_images_dir1))
copyfile_rename_labels(files_labels_dir2, path_video_combined_labels, path_video_dir2_labels, mapping)
```
