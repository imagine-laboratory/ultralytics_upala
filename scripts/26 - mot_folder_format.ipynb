```python
import os
import shutil

def transform_directory_structure(root_dir, main_dir):
    # Iterate through the root directory
    for tracker in os.listdir(root_dir):
        tracker_path = os.path.join(root_dir, tracker)
        
        if os.path.isdir(tracker_path):
            for folder in os.listdir(tracker_path):
                folder_path = os.path.join(tracker_path, folder)
                
                if os.path.isdir(folder_path):
                    # Extract the base name (e.g., DJI_0004_V_1_Aug)
                    base_name = folder.split('@')[1]
                    
                    # Create the new directory structure
                    new_base_dir = os.path.join(main_dir, base_name, 'trackers', 'mot_challenge', 'MOT15-all', tracker, 'data')
                    os.makedirs(new_base_dir, exist_ok=True)
                    
                    # Move the .txt files to the new directory
                    for file in os.listdir(folder_path):
                        if file.endswith('.txt'):
                            
                            src_file_path = os.path.join(folder_path, file)
                            dest_file_path = os.path.join(new_base_dir, file)
                            shutil.copy(src_file_path, dest_file_path)
                            print(f"Moved {src_file_path} to {dest_file_path}")

# Define the root and main directories
root_directory = 'C:/Users/dnnxl/Downloads/generated_results'
main_directory = 'C:/Users/dnnxl/Downloads/mot_challenge'
# Call the function to transform the directory structure
transform_directory_structure(root_directory, main_directory)
```

```python
import os
import shutil
from pathlib import Path
import re

def get_video_name(folder_name: str) -> str:
    """Extracts the video identifier from the folder name."""
    return folder_name.split('@')[1]

def restructure_directories(root: str, output_root: str):
    """Restructure the directory tree as specified."""
    root_path = Path(root)
    output_path = Path(output_root)
    
    for tracker in root_path.iterdir():
        if not tracker.is_dir():
            continue
        
        for sequence in tracker.iterdir():
            if not sequence.is_dir():
                continue
            print(sequence.name)
            video_name = get_video_name(sequence.name)
            new_path = output_path / video_name / "trackers/mot_challenge/MOT15-all" / tracker.name / "data"
            new_path.mkdir(parents=True, exist_ok=True)
            
            for file in sequence.iterdir():
                if file.is_file():
                    suffix = file.stem.split('_')[-2] + '_' + file.stem.split('_')[-1]


                    suffix = re.sub(r"^DJI_", "", re.sub(r"_Aug$", "", video_name))
                    text_file = "_".join(file.stem.split("_")[2:])

                    if suffix == text_file:
                        continue
                    elif "and" in suffix:
                        list_suffix = suffix.split("_and_")
                        if text_file in list_suffix:
                            continue

                    #if suffix in video_name:
                        #continue  # Skip moving if suffix is part of the video name
                    print(file.stem)
                    shutil.copy(file, new_path / file.name)
    
if __name__ == "__main__":
    root_directory = 'C:/Users/dnnxl/Downloads/generated_results/'
    output_directory = 'C:/Users/dnnxl/Downloads/mot_challenge'
    restructure_directories(root_directory, output_directory)
```
