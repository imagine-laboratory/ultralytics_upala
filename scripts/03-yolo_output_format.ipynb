```python
import os

def list_files(directory):
    try:
        # List all files in the given directory
        files = [f for f in os.listdir(directory) if os.path.isfile(os.path.join(directory, f))]
        return files
    except Exception as e:
        return str(e)  # Return the error message as a string

def list_directories(path='.'):
    return [d for d in os.listdir(path) if os.path.isdir(os.path.join(path, d))]

# Example usage:
directory_path = "C:/Users/dnnxl/Documents/GitHub/drone-sort/dataset/DJI_20240308111117_0010_V_1/valid/images"
files_in_directory = [os.path.join(directory_path, file) for file in list_files(directory_path)]
```

```python
from ultralytics import YOLO, RTDETR
# Load the weights of the YOLO model
model = YOLO("C:/Users/dnnxl/Documents/GitHub/drone-sort/weights/yolo11n/weights/best.pt")
```

```python
results = model([files_in_directory[158]], iou=0.5, conf=0.3)
```

```python
for result in results:
    for bbox in result.boxes.xyxy:
        print(bbox.cpu().detach().numpy())
```
